```python
import pandas as pd
```

```python
df = pd.read_csv("Used_bikes.csv")
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
cat_data = df.select_dtypes(include = "O")
num_data = df.select_dtypes(exclude = "O")
```

```python
cat_data
```

```python
dt = {'First Owner':4,'Second Owner':3,'Third Owner':2, 'Fourth Owner Or More':1}
```

```python
cat_data['owner'] = cat_data['owner'].map(dt)
```

```python
cat_data
```

```python
dt2 = {brand:i+1 for i,brand in enumerate(df.groupby('brand').mean().sort_values(by = 'price').index)}
```

```python
dt2
```

```python
cat_data['brand'] = cat_data['brand'].map(dt2)
```

```python
cat_data
```

```python
dt4 = {bn:i+1 for i,bn in enumerate(df.groupby('city').mean().sort_values(by = 'price').index)}
```

```python
dt4
```

```python
cat_data['city'] = cat_data['city'].map(dt4)
```

```python
cat_data = cat_data.drop(['bike_name'] , axis = 1)
```

```python
dataset = pd.concat([cat_data,num_data], axis = 1)
```

```python
x = dataset.drop('price' , axis = 1)
y = dataset[['price']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2) # 80% training 20% testing
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
prediction = lr.predict(x_test)
```

```python
y_test['prediction_price'] = prediction
y_test
```

```python
lr.score(x_train,y_train) # Train Score
```

```python
lr.score(x_test,y_test['price']) # Test Score
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse = mean_squared_error(y_test['price'] , y_test['prediction_price'])
mse
```

```python
mean_absolute_error(y_test['price'] , y_test['prediction_price'])
```

```python
import numpy as np
rmse = np.sqrt(mse)
rmse
```

```python
from sklearn.metrics import r2_score
r = r2_score(y_test['price'] , y_test['prediction_price'])
r
```
